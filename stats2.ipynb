```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
x_100 = np.linspace(-10, 10, 100)
x_10000 = np.linspace(-10, 10, 10000)
plt.figure(figsize=(12, 8))
plt.plot(x_100, stats.norm.pdf(x_100, 3, 1), label='x_100, odchylenie standardowe = 1, średnia = 3')
plt.plot(x_10000, stats.norm.pdf(x_10000, 3, 1), label='x_10000, odchylenie standardowe = 1, średnia = 3')
plt.legend(loc='upper left')
plt.show()
x_100 = stats.norm.pdf(x_100, 3, 1)
x_10000 = stats.norm.pdf(x_10000, 3, 1)
```

```python
mean_100 = np.mean(x_100)
print(f"Średnia x_100: {mean_100}")
mean_10000 = np.mean(x_10000)
print(f"Średnia x_10000: {mean_10000}")
```

```python
median_100 = np.median(x_100)
print(f"Mediana x_100: {median_100}")
median_10000 = np.median(x_10000)
print(f"Mediana x_10000: {median_10000}")
```

```python
mode_100 = stats.mode(x_100)
print(f"Moda x_100: {mode_100}")
mode_10000 = stats.mode(x_10000)
print(f"Moda x_10000: {mode_10000}")
```

```python
q0_100 = np.quantile(x_100, 0.0)
q1_100 = np.quantile(x_100, 0.25)
q3_100 = np.quantile(x_100, 0.75)
q4_100 = np.quantile(x_100, 1.0)
print(f'Q0_100: {q0_100}')
print(f'Q1_100: {q1_100}')
print(f'Q2_100: {median_100}')
print(f'Q3_100: {q3_100}')
print(f'Q4_100: {q4_100}')
q0_10000 = np.quantile(x_10000, 0.0)
q1_10000 = np.quantile(x_10000, 0.25)
q3_10000 = np.quantile(x_10000, 0.75)
q4_10000 = np.quantile(x_10000, 1.0)
print(f'Q0_10000: {q0_10000}')
print(f'Q1_10000: {q1_10000}')
print(f'Q2_10000: {median_10000}')
print(f'Q3_10000: {q3_10000}')
print(f'Q4_10000: {q4_10000}')
```

```python
variance_100 = np.var(x_100, ddof=1)
print(f'Variance_100: {variance_100}')
variance_10000 = np.var(x_10000, ddof=1)
print(f'Variance_10000: {variance_10000}')
```

```python
standard_deviation_100 = np.std(x_100, ddof=1)
print(f'Std_100: {standard_deviation_100}')
standard_deviation_10000 = np.std(x_10000, ddof=1)
print(f'Std_10000: {standard_deviation_10000}')
```

Obydwa zestawy danych posiadają praktycznie te same wartości dla statystyk opisowych. Minimalnie mniejsze odchylenie standardowe jest najlepszą miarą tego, że zbiór gęstszy jest mniej rozproszony.
